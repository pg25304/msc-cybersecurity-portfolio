```python
#unpacking a tuple
stock_price = [('Aple',200),('facebook',250),('elf',350)]
for x in stock_price:
    print(x)
```

```python
for x,y in stock_price:#unpack tuple
    print(x,0.1*y)
    
```

```python
#who is employee of the month, the one with most hours
work_hours = [('Pj',100),('Parisa', 110),('sep',120)]
def check_employee(work_hours):
    current_hour = 0
    employee_name = ''
    for name,time in work_hours:
        if time>current_hour:
            current_hour = time 
            employee_name= name
        else:
            pass
    return (employee_name , current_hour)

    
    
```

```python
w,y = check_employee(work_hours)#assign varibles to unpack tuple and use variables later as at the end of function we use RETURN
```
