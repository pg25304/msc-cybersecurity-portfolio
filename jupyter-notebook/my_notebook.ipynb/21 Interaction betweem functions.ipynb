```python
#Python provides a built-in random module, which includes a shuffle function. 
#This function shuffles a list in place, meaning it does not return a new list but modifies the original.
from random import shuffle
example= [1,2,3,4,5,6,7]
shuffle(example)#this grab a list and shuffle it and actually it does not return anything
```

```python
result = shuffle(example)#shuffle() does not return any value only shuffle the list
```

```python
def shuffle_list(mylist):
    shuffle(mylist)
    return mylist
```

```python
mylist = ['','O','']
```

```python
shuffle_list(mylist)
```

```python
def user_input():
    guess =''
    while guess not in ['0','1','2']:
         guess = input("choose one of the following numbers: from 0,1,2") #input always return string
      
    return int(guess)
        
```

```python
def check_guess(mylist,guess):
    if mylist[guess] == 'O':
        print('well done!')
        print(mylist)
    else:
        print('You are wrong')
        print(mylist)
```

```python
# When you have multiple functions that need to interact with each other, 
#you should define all the functions first, and then call them at the end of your code.
#INITIAL LIST
mylist = ['','O','']

#SHUFFLE LIST
mixedup_list = shuffle_list(mylist)
#USER GUESS
guess = user_input()
#CHECK GUESS
check_guess(mixedup_list,guess)
```
