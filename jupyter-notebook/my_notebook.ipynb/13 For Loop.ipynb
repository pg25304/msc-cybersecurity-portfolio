```python
#for loop through dictionary by default it return only keys
d = {'k1':10,'k2':23,'k3':'Parisa'}
for item in d:
    print(item)
```

```python
# you need to use  .items()
d = {'k1':10,'k2':23,'k3':'Parisa'}
for item in d.items():
    print(item)
```

```python
# you need to use  .items()
d = {'k1':10,'k2':23,'k3':'Parisa'}
for key,value in d.items(): #note that dictionary is unordered so there is no gaurented to return the result in order
    print( key)
    print(value)
    
```
