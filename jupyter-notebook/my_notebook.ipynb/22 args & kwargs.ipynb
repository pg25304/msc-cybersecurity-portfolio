```python
def myfunc(a,b):  #a and b are positional arguments 
    #return 5% of sum of a & b
    return sum((a,b))*.05
```

```python
myfunc(30,40)
```

```python
#if you may need more argument to pass but you do not know yet ONE WAY is define them as zero
def myfunc(a,b,c=0,d=0):
    return sum((a,b,c,d))*0.05 #now if you call it same as before as follows you receive the same result
myfunc(30,40)
```

```python
#second way to use args keyword
#If you print out args inside the function, you will see that it is just a tuple. 
#The star or asterisk in the parameter tells Python to collect all positional arguments into a tuple.
#star args return tuple
def myfunc(*args):
    print(args)
    return sum(args)*0.05
#By convention, the name args is used, but technically, any name can be used after the asterisk. For example, 
#*spam would work the same way, but it is recommended to use args for clarity and consistency.
```

```python
myfunc(30,40)
```

```python
#you can now pass more parameters:
myfunc(10,20,30,40)
```

```python
#star star kwargs (**kwargs)return dictionary
def myfunc(**kwargs):
    print(kwargs)
    if 'fruit' in kwargs:
        #print("my choice of fruity is:  {}".format(kwargs[ 'fruit']))
        print(f"the fruity is {kwargs['fruit']}")
    else:
        print('I did not find any fruit here')
```

```python
myfunc( fruit='apple', vegi = 'lettuce')
```

```python
def myfunc(*args,**kwargs):
    print(args)
    print(kwargs)
    print(f"I have {args[0]} {kwargs['fruit']}")#when you call function parameter should ame order start with args then with kwargs
```

```python
myfunc(10,2,3,4, kir='kos',fruit='apple')
```
