```python
celcius = [10,20,30,40,50]
Fahrenheit = [((9/5)+temp+32) for temp in celcius] # consider (9/5)+temp+32) as xin x for x in...
```

```python
Fahrenheit
```

```python
Fahrenheit=[]# this is exactly same as above codes
celcius = [10,20,30,40,50]
for temp in celcius:
    Fahrenheit.append((9/5)+temp+32)
```

```python
#you can write if and else in one line in list comprehension but readability is really important
#especially when come back later
result = [ x if x%2==0 else 'ODD' for x in range(1,20)]
result
```

```python
mylist = []#NESTED LOOP
for x in [1,2,3,4,5]:
    for y in [2,4,6]:
        mylist.append(x*y)
```

```python
mylist
```

```python
#List comprehension versio
mylist = [x*y for x in [1,2,3,4,5] for y in [2,4,6]]
mylist
```

```python
ylist = [((x+y)-z) for x in [1,2] for y in[5,6] for z in [1,2] if x+y <= 7] #NOTE you cannot use else after if in list comprehension
ylist
```
