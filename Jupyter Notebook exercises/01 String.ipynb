```python
#Strng formatting for printing(string interpolation = refers to the process of inserting something into an existing sequence or structure.)
#string.format() syntax:"string here {place holder} {place holder 2}".format('something1','sth2') 
#print("This is a {} {} {}".format('fast','gray','cat'))  # Using string.format() with empty braces {} as placeholders that will be filled in order with the arguments 'fast', 'gray', 'cat'
print("This is a {} {} {}".format('fast','gray','cat'))
print("This is a {0} {2} {1}".format('fast','gray','cat'))
```

```python
#In Python, float formatting using {placeholder:width.precision} allows you to control the width and decimal precision of a floating-point number when displaying it
result = 100/110
result
```

```python
print('This is {r:5.3f}'.format(r=result))#print('string {placeholder:width.precision}'.format(provide value for key r)
```
