# Assistant
Here are several examples of string formatting for the `print` function in Python:


```python
# Basic string formatting using .format()
name = "Alice"
age = 30
print("My name is {} and I am {} years old.".format(name, age))

# Using positional arguments
print("My name is {0} and I am {1} years old.".format(name, age))

# Using named arguments
print("My name is {n} and I am {a} years old.".format(n=name, a=age))

# Formatting numbers
pi = 3.14159
print("Pi rounded to 2 decimal places: {:.2f}".format(pi))
print(f"Pi rounded to 2 decimal places:  {pi:10.3f}")

# Width and alignment
print("Right aligned: {:>50}".format("text"))  # Width of 10,space added to the left
print("Left aligned: {:<10}".format("text"))   # space added to the right
print("Centered: {:^10}".format("text"))       # Width of 10, centered

# Formatting with f-strings (Python 3.6+)
print(f"My name is {name} and I am {age} years old.")
print(f"Pi rounded to 2 decimal places:{pi:10.2f}")
print(f"Right aligned: {pi:>10.2f} world")

# Padding numbers with zeros
print("Padded with zeros: {:05d}".format(42))  # 00042

# Including the sign
print("With sign: {:+d}".format(42))  # +42
print("With sign: {:+d}".format(-42)) # -42

# Thousands separator
print("With commas: {:,}".format(1000000))  # 1,000,000
```


These examples cover most common string formatting scenarios you might encounter when using the `print` function.
