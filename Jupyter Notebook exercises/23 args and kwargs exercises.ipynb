```python
#unctions #9: pick evens
#Define a function called myfunc that takes in an arbitrary number of arguments, and returns a list containing only those 
#arguments that are even.Remember, don't run the function, simply provide the definition.
#To give an idea what the function would look like when tested:myfunc(5,6,7,8)# Output: [6, 8]
#Added note: this exercise requires that the function return a list. Print statements will not work here.
def myfunc(*args):
    # Initialize an empty list to store even numbers
    x = []
    
    # Loop through all arguments
    for item in args:
        # Check if the number is even
        if item % 2 == 0:
            # Add even numbers to the list
            x.append(item)
    
    # Return the list of even numbers
    return x

# Test the function
myfunc(1, 2, 4, 5, 7, 6, 8, 0)
```

```python
#Functions #10: skyline
#Define a function called myfunc that takes in a string, and returns a matching string where every even letter is uppercase,
#and every odd letter is lowercase. Assume that the incoming string only contains letters, and don't worry about numbers,
#spaces or punctuation. The output string can start with either an uppercase or lowercase letter, so long as letters
#alternate throughout the string.
def myfunc(*args):
    # Check if any arguments were provided
    if not args:
        return ""
    
    # Take the first argument as the string to process
    string = args[0]
    
    # Initialize an empty list to store modified letters
    x = []
    i = 0
    
    # Loop through each letter in the string
    for letter in string:
        i += 1
        if i % 2 == 0:
            # Make even-positioned letters uppercase
            x.append(letter.upper())
        else:
            # Make odd-positioned letters lowercase
            x.append(letter.lower())
    
    # Join the list elements back into a string and return
    return ''.join(x)
 
# Example usage
# myfunc("hello") would return "hElLo"
```

```python
myfunc()
```
