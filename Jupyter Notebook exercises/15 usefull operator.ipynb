```python
#enumerate() function willindex count automatically in form of tuple
word = 'abcde'
#word = [1,2,3,4,5,6] this works either
for item in enumerate(word):
    print(item)
```

```python
#enumerate() function willindex count automatically in form of tuple
word = 'abcde'
#word = [1,2,3,4,5,6] this works either
for index,letter in enumerate(word):
    print(index)
    print(letter)
    print('\n')
```

```python
#when use zip() function it does opposite of enumerate() means combine data if there are several uneven list it match it 
#up to the number of shortest list number
list1=[1,2,3,4,5,6]
list2=['a','b','c','d','e','f','g','h']
list3=[100,200,300]
for item in zip(list1,list2,list3):
    print(item)
```

```python
list1=[1,2,3,4,5,6]
list2=['a','b','c','d','e','f','g','h']
list3=[100,200,300]
for item1,item2,item3 in zip(list1,list2,list3):
    print(item1)
    print(item2)
    print(item3)
    print('\n')
```
