```python
# Importing Fernet module from Cryptography library,Fernet class is implementation of symmetric authenticated cryptography
# pip install cryptography - already installed
from cryptography.fernet import Fernet
# Step 1: defining function generate_key() to Generate and save a key
def generate_key():    
    key = Fernet.generate_key()  # Create a new symmetric key
    with open("filekey.key", "wb") as key_file: # open a file name filekey.key in write binary mode
        key_file.write(key)  # Save the key to a file
        
#step2 function to read and load the key from the file
def load_key():
    return open("filekey.key", "rb").read()  # Read the key from the file

# STep3 Function to encrypt the file using Fernet class from cryptography
def encrypt_file(filename, key): # -filename- is the path of the file to be encrypted
    # create a Fernet object using the keyto use for encryption and decryption
    fernet = Fernet(key) 
    #Opens the file specified by filename in read-binary mode (rb) and assigns the file object to the variable file
    with open(filename, "rb") as file:
        #Reads the content of the opened file and assigns it to the original variable.
        original = file.read()
    #Encrypts the original file content using the Fernet object's encrypt() method
    #The encrypted data is stored in the encrypted variable.
    encrypted = fernet.encrypt(original)
    #Opens the same file specified by filename in write-binary and assign the file object to encrypted_file
    with open(filename, "wb") as encrypted_file:
        #Writes the encrypted data to the file, effectively overwriting the original content.
        encrypted_file.write(encrypted)  

#  Step 4: defining decrypt_file function to decrypt the file
# -filename- is the path of the file to be encrypted
def decrypt_file(filename, key):
    #Creates a Fernet object using the provided key. This object will be used for decryption
    fernet = Fernet(key)
    #Opens the file specified by filename in read-binary mode (rb) and assigns the file object to the variable enc_file.
    with open(filename, "rb") as enc_file:
        #Reads the content of the opened encrypted file and assigns it to the encrypted variable.
        encrypted = enc_file.read()
    #Decrypts the encrypted file content using the Fernet object's decrypt() method. The decrypted data is stored in the decrypted variable.
    decrypted = fernet.decrypt(encrypted)
    #Opens the same file specified by filename in write-binary mode (wb) and assigns the file object to the variable dec_file.
    with open(filename, "wb") as dec_file:
        #Writes the decrypted data to the file, Overwriting the encrypted content with the original content.
        dec_file.write(decrypted)






```

```python
#Creating a text file for encryption and decryption
with open("test.txt", "w") as f:
    f.write("This is a secret message for encryption testing.Payman is going to carry on this test f. ")
```

```python
#Test the Utility
generate_key()
key = load_key()
print(key)
```

```python
#Show the file before encryption
with open("test.txt", "r") as f:
    print("Before encryption:", f.read())
```

```python
encrypt_file("test.txt", key)
with open("test.txt", "r") as f:
    print("After encryption:", f.read())
```

```python
decrypt_file("test.txt", key)
with open("test.txt", "r") as f:
    print("After decryption:", f.read())
```

```python
#Measure the time taken for encryption & decryption
%time encrypt_file("test.txt", key)
%time decrypt_file("test.txt", key)
```
